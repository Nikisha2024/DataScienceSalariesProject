```python
# changing values of df column 'experience_level'
df['experience_level'] = df['experience_level'].replace({
    'SE': 'Senior Level/Expert',
    'MI': 'Medium Level/Intermediate',
    'EN': 'Entry Level',
    'EX': 'Executive Level'
})

df['experience_level']
```

```python

def summary_statistics(data):
    # Calculate sum, mean, standard deviation, skewness, and kurtosis
    sum_val = np.sum(data)
    mean_val = np.mean(data)
    std_dev = np.std(data)
    skewness = skew(data)
    kurt = kurtosis(data)

    # Store the summary statistics in a dictionary
    summary = {
        "Sum": sum_val,
        "Mean": mean_val,
        "Standard Deviation": std_dev,
        "Skewness": skewness,
        "Kurtosis": kurt
    }
    
    return summary


```

```python
# calling the method
data = df['salary_in_usd']
result = summary_statistics(data)


# Print the summary statistics
print('All the summary statistics are')
for key, value in result.items():
    print(key + ":", value)
```

```python
# calculating top 15 jobs from values_counts
top_job=df['job_title'].value_counts().head(15).reset_index() 
top_job
```

```python
spec = top_job['job_title']
df_filtered = df[df['job_title'].isin(spec)] #filtering df dataframe job_title with spec dataframe
#grouping average salary according to job title
average_income_top_jobs = df_filtered.groupby('job_title')['salary_in_usd'].mean()
average_income_top_jobs
```

```python
plt.figure(figsize=(10, 6))
average_income_top_jobs.sort_values(ascending=False).plot(kind='bar', color='skyblue')
plt.title('Average Salary for Top 15 Jobs') # gives title to the diagram
plt.xlabel('Job Title') # it represents group of data present in x-label
plt.ylabel('Average Salary ($)') # it represents group of data present in y-label
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
plt.tight_layout()  # Adjusting layout to prevent clipping of labels
plt.show()
```
